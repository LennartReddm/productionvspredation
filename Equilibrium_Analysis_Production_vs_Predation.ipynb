```python
# Load packages

from matplotlib import *
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import *
import numpy as np
import pandas as pd
import quantecon
import quantecon.game_theory as gt
import ipywidgets as widgets

# Set numpy print width
np.set_printoptions(linewidth=300)
np.set_printoptions(precision=3, suppress=False)

pd.options.mode.chained_assignment = None  # default='warn'

# %load_ext google.colab.data_table
```

```python
# FUNCTIONS


## PAY-OFF FUNCTIONS

# Defender conflict pay-off function for the PAD-C.
def poDef(a, d, TDef, eDef, kDef):  
    if a > d:
        return 0
    else:
        return (eDef - d + lDef(d, TDef, eDef, kDef))

# Defender production function
def lDef(d, TDef, eDef, kDef):
    if (eDef - d) >= TDef:
        return kDef - TDef
    else:
        return 0

# Attacker conflict pay-off function for the PAD-C.
def poAtt(a, d, TAtt, eAtt, TDef, eDef, kAtt, kDef):  
    if a > d:
        return (eAtt - a + lAtt(a, TAtt, eAtt, kAtt)) + (eDef - d + lDef(d, TDef, eDef, kDef))
    else:
        return (eAtt - a + lAtt(a, TAtt, eAtt, kAtt))

# Attacker production function
def lAtt(a, TAtt, eAtt, kAtt):
    if (eAtt - a) >= TAtt:
        return kAtt - TAtt
    else:
        return 0



## OTHER FUNCTIONS


### Create pay-off matrizes

# Function to create Attacker Pay-off Matrix
def createAttPOMat(eDef, eAtt, TDef, TAtt, kDef, kAtt):
    POAmat = np.zeros((eAtt+1, eDef+1),dtype=int)
    for i in range(eAtt+1):
        for j in range(eDef+1):
            POAmat[i,j]=poAtt(i, j, TAtt, eAtt, TDef, eDef, kAtt, kDef)
    return POAmat

# Function to create Defender Pay-off Matrix
def createDefPOMat(eDef, eAtt, TDef, kDef):
    PODmat = np.zeros((eDef+1,eAtt+1),dtype=int)
    for i in range(eDef+1):
        for j in range(eAtt+1):
            PODmat[i,j]=poDef(i, j, TDef, eDef, kDef)
    return PODmat

# Function to create an array with pay-offs for each a and d, and T and k as parameters, for Attacker
def PADCMatAtt(T, k):
  TDef = T
  TAtt = T
  kDef = k
  kAtt = k
  AttPOMat = createAttPOMat(eDef, eAtt, TDef, TAtt, kDef, kAtt)
  #AttPOMat = gt.NormalFormGame(AttPOMat)
  return AttPOMat

# Wrapper Function to create a dynamic array with pay-offs for each a and d, and T and k as parameters, for Attacker
def PADCMatrixAtt():
    PADCMatrixAtt = widgets.interact(PADCMatAtt,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                  k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return PADCMatrixAtt

# Function to create an array with pay-offs for each a and d, and T and k as parameters, for defender
def PADCMatDef(T, k):
  TDef = T
  kDef = k
  DefPOMat = createDefPOMat(eDef, eAtt, TDef, kDef)
  return DefPOMat

# Wrapper Function to create a dynamic array with pay-offs for each a and d, and T and k as parameters, for Defender
def PADCMatrixDef():
    PADCMatrixDef = widgets.interact(PADCMatDef,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                  k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return PADCMatrixDef



### Calculate and return equilibrium predictions based on pay-off matrizes using Lemke-Howson algorithm 

# Function that created equlibrium outputs for each T and k.
def PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt):
    PODmat = createDefPOMat(eDef, eAtt, TDef, kDef)
    POAmat = createAttPOMat(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    POADmat = np.dstack((POAmat,PODmat))
    g_POADmat = gt.NormalFormGame(POADmat)
    equilib = gt.lemke_howson(g_POADmat)
    return equilib



### Equilibrium tables

# Function to create a table with equilibrium predictions with the same production pay-off and treshold for attacker and defender
def PADCEquilibTable(T, k):
  TDef = T
  TAtt = T
  kDef = k
  kAtt = k
  # Create table with equilibrium strategies
  PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
  PADCEquilibTab = pd.DataFrame({'Strategy': list(range(0, eAtt+1)),
                               'Att Probability': PADCEquilibArr[0],
                               'Def Probability': PADCEquilibArr[1]})
  return PADCEquilibTab

# Wrapper Function to create a dynamic table with equilibrium predictions with the same production pay-off and treshold
def PADCEquilibTab():
    PADCEquilibTab = widgets.interact(PADCEquilibTable, T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                        k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return PADCEquilibTab


# Function to create a dynamic table with equilibrium predictions for separate Attacker and Defender production pay-offs and tresholds
def PADCEquilibTableSep(TDef, TAtt, kDef, kAtt):
  # Create table with equilibrium strategies
  PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
  PADCEquilibTab = pd.DataFrame({'Strategy': list(range(0, eAtt+1)),
                               'Att Probability': PADCEquilibArr[0],
                               'Def Probability': PADCEquilibArr[1]})
  return PADCEquilibTab

# Table calculation optimal choice between investment in production p and Keep for configuration of Threshold T and pay-off k.
# For each level of conflict investment (here called CI) there is only *one* optimal choice between investment in production p and Keep. 
def tableOptimalProdPO(T, k, e):
  PADCOptimalp = pd.DataFrame({'Conflict Investment (CI) level': list(range(0, e+1))})
  PADCOptimalp['MU left'] = list(reversed(range(0, e+1)))
  PADCOptimalp['Threshold T'] = T
  PADCOptimalp['Optimal Keep for CI level'] = NAN
  PADCOptimalp['Optimal p for CI level'] = NAN
  PADCOptimalp['Max PO prod + keep for CI level'] = NAN
  for index, row in PADCOptimalp.iterrows():
    i = row['MU left']
    if i >= T:
      PADCOptimalp['Optimal p for CI level'][index] = T
      PADCOptimalp['Optimal Keep for CI level'][index] = PADCOptimalp['MU left'][index] - T
      PADCOptimalp['Max PO prod + keep for CI level'][index] = k + PADCOptimalp['MU left'][index] - T
    else:
      PADCOptimalp['Optimal p for CI level'][index] = 0
      PADCOptimalp['Optimal Keep for CI level'][index] = PADCOptimalp['MU left'][index]    
      PADCOptimalp['Max PO prod + keep for CI level'][index] = PADCOptimalp['MU left'][index] 
  return PADCOptimalp 



### Equilibrium plots

# Function to create a dynamic plot with with equilibrium predictions with the same production pay-off and treshold for attacker and defender.
def PADCEquilibPlot(T, k):
  TDef = T
  TAtt = T
  kDef = k
  kAtt = k
  # Create table with equilibrium strategies
  PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
  PADCEquilibTable = pd.DataFrame({'Strategy': list(range(0, eAtt+1)),
                               'Att Probability': PADCEquilibArr[0],
                               'Def Probability': PADCEquilibArr[1]})
  # Create plot with equilibrium strategies
  fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,5))  # 1 row, 2 columns
  PADCEquilibTable.plot.bar(x='Strategy', y=['Att Probability'], color=['Red'], ylim=(0,1), width=0.8, rot=50, ax=ax1)
  PADCEquilibTable.plot.bar(x='Strategy', y=['Def Probability'], color=['Blue'], ylim=(0,1), width=0.8, rot=50, ax=ax2)
  plt.suptitle('Equilibrium play for attacker and defender in the PAD-C', verticalalignment='bottom')
  ax2.set_yticklabels([],minor=False)
  plt.tight_layout()  # Optional ... often improves the lay

# Function to create a dynamic plot with equilibrium predictions for separate Attacker and Defender pay-offs and tresholds
def PADCEquilibPlotSep(TDef, TAtt, kDef, kAtt):
  # Create table with equilibrium strategies
  PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
  PADCEquilibTable = pd.DataFrame({'Strategy': list(range(0, eAtt+1)),
                               'Att Probability': PADCEquilibArr[0],
                               'Def Probability': PADCEquilibArr[1]})
  # Create plot with equilibrium strategies
  fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,5))  # 1 row, 2 columns
  PADCEquilibTable.plot.bar(x='Strategy', y=['Att Probability'], color=['Red'], ylim=(0,1), width=0.8, rot=50, ax=ax1)
  PADCEquilibTable.plot.bar(x='Strategy', y=['Def Probability'], color=['Blue'], ylim=(0,1), width=0.8, rot=50, ax=ax2)
  plt.suptitle('Equilibrium play for attacker and defender in the PAD-C', verticalalignment='bottom')
  ax2.set_yticklabels([],minor=False)
  plt.tight_layout()  # Optional ... often improves the lay

# Wrapper Function to create a dynamic plot with equilibrium predictions with the same production pay-off and treshold
def PADCEquilibPlotInt():
    PADCEquilibPlotInt = widgets.interactive(PADCEquilibPlot, {'manual': False}, T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                                                 k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return PADCEquilibPlotInt



## Functions that compute additional variables based on equilibrium predictions


### Average Conflict investment
def avgCInvAtt(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    PADCEquilibTable = pd.DataFrame({'Strategy': list(range(0, eAtt+1)),
                               'Att Probability': PADCEquilibArr[0],
                               'Def Probability': PADCEquilibArr[1]})
    # Calculate weighted average
    d = PADCEquilibTable['Strategy']
    w = PADCEquilibTable['Att Probability']
    return (d * w).sum()

def averageConflictInvestmentAtt():
    avgConflictInvestmentDef = widgets.interact(avgCInvAtt,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                             k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return avgConflictInvestmentDef

def avgCInvDef(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    PADCEquilibTable = pd.DataFrame({'Strategy': list(range(0, eAtt+1)),
                               'Att Probability': PADCEquilibArr[0],
                               'Def Probability': PADCEquilibArr[1]})
    # Calculate weighted average
    d = PADCEquilibTable['Strategy']
    w = PADCEquilibTable['Def Probability']
    return (d * w).sum()

def averageConflictInvestmentDef():
    avgConflictInvestmentDef = widgets.interact(avgCInvDef,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                             k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))


### Average Conflict Force

# A function to give the conflict force under equilibrium. 
# Tested against the predictions on "Force" from the 2015 paper, works!
def invForce(df, str, p_weight):
    df = df.drop(0)
    strategies = df[str]
    pweights = df[p_weight]
    return (strategies * pweights).sum() / pweights.sum()

def invForceAtt(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    PADCEquilibTable = pd.DataFrame({'Conflict Strategy': list(range(0, eAtt+1)),
                                     'Att Probability': PADCEquilibArr[0],
                                     'Def Probability': PADCEquilibArr[1]})
    force = invForce(PADCEquilibTable, 'Conflict Strategy', 'Att Probability')
    return force

def invForceDef(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    PADCEquilibTable = pd.DataFrame({'Conflict Strategy': list(range(0, eAtt+1)),
                                     'Att Probability': PADCEquilibArr[0],
                                     'Def Probability': PADCEquilibArr[1]})
    force = invForce(PADCEquilibTable, 'Conflict Strategy', 'Def Probability')
    return force

def averageConflictForceAtt():
    averageConflictForceAtt = widgets.interact(invForceAtt,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                               k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return averageConflictForceAtt


def averageConflictForceDef():
    averageConflictForceDef = widgets.interact(invForceDef,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                               k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return averageConflictForceDef


### Average Conflict Frequency

# A function that returns the investment frequency.
def invFreq(df, str, p_weight):
    d = df.loc[df[str] == 0, p_weight]
    d = float(d)
    return 1 - d

def invFreqAtt(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    PADCEquilibTable = pd.DataFrame({'Conflict Strategy': list(range(0, eAtt+1)),
                                     'Att Probability': PADCEquilibArr[0],
                                     'Def Probability': PADCEquilibArr[1]})
    freq = invFreq(PADCEquilibTable, 'Conflict Strategy', 'Att Probability')
    return freq

def invFreqDef(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    PADCEquilibTable = pd.DataFrame({'Conflict Strategy': list(range(0, eAtt+1)),
                                     'Att Probability': PADCEquilibArr[0],
                                     'Def Probability': PADCEquilibArr[1]})
    freq = invFreq(PADCEquilibTable, 'Conflict Strategy', 'Def Probability')
    return freq

def averageConflictFreqAtt():
    averageConflictFreqAtt = widgets.interact(invFreqAtt,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                               k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return averageConflictFreqAtt

def averageConflictFreqDef():
    averageConflictFreqDef = widgets.interact(invFreqDef,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                           k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return averageConflictFreqDef


### Average Production Investment

def avgPInvAtt(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    tblOptimalProdPO = tableOptimalProdPO(T, k, e)
    PADCEquilibTable = pd.DataFrame({'Conflict Strategy': list(range(0, eAtt+1)),
                                     'Optimal production strategy': tblOptimalProdPO['Optimal p for CI level'],
                                     'Att Probability': PADCEquilibArr[0],
                                     'Def Probability': PADCEquilibArr[1]})
    # Calculate weighted average
    d = PADCEquilibTable['Optimal production strategy']
    w = PADCEquilibTable['Att Probability']
    return (d * w).sum()

def averageProductionInvestmentAtt():
    avgProductionInvestmentAtt = widgets.interact(avgPInvAtt,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                               k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return avgProductionInvestmentAtt

def avgPInvDef(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    tblOptimalProdPO = tableOptimalProdPO(T, k, e)
    PADCEquilibTable = pd.DataFrame({'Conflict Strategy': list(range(0, eAtt+1)),
                                     'Optimal production strategy': tblOptimalProdPO['Optimal p for CI level'],
                                     'Att Probability': PADCEquilibArr[0],
                                     'Def Probability': PADCEquilibArr[1]})
    # Calculate weighted average
    d = PADCEquilibTable['Optimal production strategy']
    w = PADCEquilibTable['Def Probability']
    return (d * w).sum()

def averageProductionInvestmentDef():
    avgProductionInvestmentDef = widgets.interact(avgPInvDef,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                               k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return avgProductionInvestmentDef


### Average KEEP
def avgKeepAtt(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    tblOptimalProdPO = tableOptimalProdPO(T, k, e)
    PADCEquilibTable = pd.DataFrame({'Conflict Strategy': list(range(0, eAtt+1)),
                                     'Optimal Keep': tblOptimalProdPO['Optimal Keep for CI level'],
                                     'Att Probability': PADCEquilibArr[0],
                                     'Def Probability': PADCEquilibArr[1]})
    # Calculate weighted average
    d = PADCEquilibTable['Optimal Keep']
    w = PADCEquilibTable['Att Probability']
    return (d * w).sum()

def averageKeepAtt():
    averageKeepAtt = widgets.interact(avgKeepAtt,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                   k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return averageKeepAtt

def avgKeepDef(T, k):
    TDef = T
    TAtt = T
    kDef = k
    kAtt = k
    # Create table with equilibrium strategies
    PADCEquilibArr = PADCequilib(eDef, eAtt, TDef, TAtt, kDef, kAtt)
    tblOptimalProdPO = tableOptimalProdPO(T, k, e)
    PADCEquilibTable = pd.DataFrame({'Conflict Strategy': list(range(0, eAtt+1)),
                                     'Optimal Keep': tblOptimalProdPO['Optimal Keep for CI level']})
    # Calculate weighted average
    d = PADCEquilibTable['Optimal Keep']
    w = PADCEquilibTable['Def Probability']
    return (d * w).sum()

def averageKeepDef():
    averageKeepDef = widgets.interact(avgKeepDef,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                   k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
    return averageKeepDef


### Average Expected Earnings

def expEarningsAtt(T, k):
  TDef = T
  TAtt = T
  kDef = k
  kAtt = k

  PODmat = createDefPOMat(eDef, eAtt, TDef, kDef)
  POAmat = createAttPOMat(eDef, eAtt, TDef, TAtt, kDef, kAtt)
  POADmat = np.dstack((POAmat,PODmat))
  g_POADmat = gt.NormalFormGame(POADmat)
  equilib = gt.lemke_howson(g_POADmat)

  probPOMat = np.empty((eDef+1, eAtt+1))
  for i in range(eDef+1):
    for j in range(eAtt+1):
      # print(equilib[0][i], equilib[1][j], equilib[0][i] * equilib[1][j])
      probPOMat[i,j] = equilib[0][i] * equilib[1][j]
  expEarningsAtt = np.multiply(probPOMat, POAmat).sum()
  return expEarningsAtt

def expEarningsDef(T, k):
  TDef = T
  TAtt = T
  kDef = k
  kAtt = k

  PODmat = createDefPOMat(eDef, eAtt, TDef, kDef)
  POAmat = createAttPOMat(eDef, eAtt, TDef, TAtt, kDef, kAtt)
  POADmat = np.dstack((POAmat,PODmat))
  g_POADmat = gt.NormalFormGame(POADmat)
  equilib = gt.lemke_howson(g_POADmat)

  probPOMat = np.empty((eDef+1, eAtt+1))
  for i in range(eDef+1):
    for j in range(eAtt+1):
      # print(equilib[0][i], equilib[1][j], equilib[0][i] * equilib[1][j])
      probPOMat[i,j] = equilib[0][i] * equilib[1][j]
  expEarningsDef = np.multiply(probPOMat, PODmat).sum()
  return expEarningsDef

def expEarningsAttInt():
  expEarningsAttInt = widgets.interact(expEarningsAtt,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                        k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
  return expEarningsAttInt

def expEarningsDefInt():
  expEarningsDefInt = widgets.interact(expEarningsDef,  T=widgets.IntSlider(min=0, max=eAtt, step=1, value=0), 
                                                        k=widgets.IntSlider(min=0, max=kMax, step=1, value=0))
  return expEarningsDefInt


## Heatmap functions


### Heatmap function for two heatmaps with attacker and defender average prediction
def AvgsUnderEquilibHMAx(e, kMax, sclEndBtm, sclEndTop, avgFuncAtt, avgFuncDef, title, axs):
  T = np.arange(0, e+1)
  k = np.arange(0, kMax+1)

  # filling an array for the attacker heatmap, value by value
  arrayAvgsAtt = numpy.empty((T.size, k.size))
  for i in range(T.size):
    for j in range(k.size):
        if j < i: # If price k is smaller than treshold T there is no reason to invest in production. Player fall back on the keep option to maximise their PO and the game reduces to the AD-C.
          arrayAvgsAtt[i,j] = avgFuncAtt(0, 0)
        else:
          arrayAvgsAtt[i,j] = avgFuncAtt(T[i], k[j])
  arrayAvgsAtt = transpose(arrayAvgsAtt) # Transpose in order to plot T on x-axis and k on y-axis.

  # filling an array for the defender heatmap, value by value
  arrayAvgsDef = numpy.empty((T.size, k.size))
  for i in range(T.size):
    for j in range(k.size):
        if j < i: # If price k is smaller than treshold T there is no reason to invest in production. Player fall back on the keep option to maximise their PO and the game reduces to the AD-C.
          arrayAvgsDef[i,j] = avgFuncDef(0, 0)
        else:
          arrayAvgsDef[i,j] = avgFuncDef(T[i], k[j])
  arrayAvgsDef = transpose(arrayAvgsDef) # Transpose in order to plot T on x-axis and k on y-axis.

  # Create the heatmaps plots
  # f,(ax1,ax2, axcb) = plt.subplots(1,3, figsize=(11,5), gridspec_kw={'width_ratios':[1,1,0.04]})
  axs[0].get_shared_y_axes().join(axs[1])

  norm = plt.Normalize(sclEndBtm, sclEndTop)

  g1 = sns.heatmap(arrayAvgsAtt, cmap="YlOrRd",cbar=False, ax=axs[0], norm=norm)
  # g1.set_ylabel('Production pay-off k')
  # g1.set_xlabel('Production threshold T') 
  g1.set_xlabel('') 
  g1.invert_yaxis()
  g2 = sns.heatmap(arrayAvgsDef,cmap="YlOrRd", cbar=True, ax=axs[1], cbar_ax=axs[2], norm=norm) 
  g2.set_ylabel('')
  g2.set_xlabel('')
  g2.set_yticks([])
  g2.invert_yaxis()
  
  g1.tick_params(axis='x', labelsize=16, rotation= 0)
  g2.tick_params(axis='x', labelsize=16, rotation= 0)
  g1.tick_params(axis='y', labelsize=16, rotation= 0)
  axs[2].tick_params(labelsize=16)

  g1.set_yticks(np.arange(0, 41, step=5))
  g1.set_yticklabels(np.arange(0, 41, step=5))
  g1.set_xticks(np.arange(0, 21, step=2))
  g1.set_xticklabels(np.arange(0, 20, step=2))

  g2.set_xticks(np.arange(0, 21, step=2))
  g2.set_xticklabels(np.arange(0, 21, step=2))


### Heatmap function with one heatmap showing the difference between attacker and defender predictions (e.g. for inequality between them)
def AvgsUnderEquilibDiffHMAx(e, kMax, sclEndBtm, sclEndTop, avgFuncAtt, avgFuncDef, title, axs):
  T = np.arange(0, e+1)
  k = np.arange(0, kMax+1)

  # filling an array for the attacker heatmap, value by value
  arrayAvgs = numpy.empty((T.size, k.size))
  for i in range(T.size):
    for j in range(k.size):
        if j < i: # If price k is smaller than treshold T there is no reason to invest in production. Player fall back on the keep option to maximise their PO and the game reduces to the AD-C.
          arrayAvgs[i,j] = avgFuncAtt(0, 0) - avgFuncDef(0, 0)
        else:
          arrayAvgs[i,j] = avgFuncAtt(T[i], k[j]) - avgFuncDef(T[i], k[j])
  arrayAvgs = transpose(arrayAvgs) # Transpose in order to plot T on x-axis and k on y-axis.

  # Create the heatmap
  norm = plt.Normalize(sclEndBtm, sclEndTop)

  g1 = sns.heatmap(arrayAvgs, cmap="YlOrRd",cbar=True, ax=axs[0], cbar_ax=axs[1], norm=norm)
  g1.set_ylabel('Production pay-off k')
  g1.set_xlabel('Production threshold T') 
  g1.invert_yaxis()

  g1.tick_params(axis='x', labelsize=16, rotation= 0)
  g1.tick_params(axis='y', labelsize=16, rotation= 0)
  axs[1].tick_params(labelsize=16)

  g1.set_yticks(np.arange(0, 40, step=5))
  g1.set_yticklabels(np.arange(0, 41, step=5))
  g1.set_xticks(np.arange(0, 20, step=2))
  g1.set_xticklabels(np.arange(0, 21, step=2))


### Heatmap function for two heatmaps with attacker and defender average prediction
def AvgsUnderEquilibHM(e, kMax, sclEndBtm, sclEndTop, avgFuncAtt, avgFuncDef, title):
  T = np.arange(0, e+1)
  k = np.arange(0, kMax+1)

  # filling an array for the attacker heatmap, value by value
  arrayAvgsAtt = numpy.empty((T.size, k.size))
  for i in range(T.size):
    for j in range(k.size):
        if j < i: # If price k is smaller than treshold T there is no reason to invest in production. Player fall back on the keep option to maximise their PO and the game reduces to the AD-C.
          arrayAvgsAtt[i,j] = avgFuncAtt(0, 0)
        else:
          arrayAvgsAtt[i,j] = avgFuncAtt(T[i], k[j])
  arrayAvgsAtt = transpose(arrayAvgsAtt) # Transpose in order to plot T on x-axis and k on y-axis.

  # filling an array for the defender heatmap, value by value
  arrayAvgsDef = numpy.empty((T.size, k.size))
  for i in range(T.size):
    for j in range(k.size):
        if j < i: # If price k is smaller than treshold T there is no reason to invest in production. Player fall back on the keep option to maximise their PO and the game reduces to the AD-C.
          arrayAvgsDef[i,j] = avgFuncDef(0, 0)
        else:
          arrayAvgsDef[i,j] = avgFuncDef(T[i], k[j])
  arrayAvgsDef = transpose(arrayAvgsDef) # Transpose in order to plot T on x-axis and k on y-axis.

  # Create the heatmaps plots
  f,(ax1,ax2, axcb) = plt.subplots(1,3, figsize=(11,5), gridspec_kw={'width_ratios':[1,1,0.04]})
  ax1.get_shared_y_axes().join(ax2)

  norm = plt.Normalize(sclEndBtm, sclEndTop)

  g1 = sns.heatmap(arrayAvgsAtt, cmap="YlOrRd",cbar=False, ax=ax1, norm=norm)
  g1.set_ylabel('Production pay-off k')
  g1.set_xlabel('Production threshold T') 
  g1.invert_yaxis()
  g2 = sns.heatmap(arrayAvgsDef,cmap="YlOrRd", cbar=True, ax=ax2, norm=norm, cbar_ax=axcb) 
  g2.set_ylabel('')
  g2.set_xlabel('')
  g2.set_yticks([])
  g2.invert_yaxis()

  g1.tick_params(axis='x', labelsize=9, rotation= 50)
  g2.tick_params(axis='x', labelsize=9, rotation= 50)
  g1.tick_params(axis='y', labelsize=9, rotation= 0)

  plt.suptitle(title, verticalalignment='bottom')
  plt.tight_layout()


### Heatmap function with one heatmap showing the difference between attacker and defender predictions (e.g. for inequality between them)
def AvgsUnderEquilibDiffHM(e, kMax, sclEndBtm, sclEndTop, avgFuncAtt, avgFuncDef, title):
  T = np.arange(0, e+1)
  k = np.arange(0, kMax+1)

  # filling an array for the attacker heatmap, value by value
  arrayAvgs = numpy.empty((T.size, k.size))
  for i in range(T.size):
    for j in range(k.size):
        if j < i: # If price k is smaller than treshold T there is no reason to invest in production. Player fall back on the keep option to maximise their PO and the game reduces to the AD-C.
          arrayAvgs[i,j] = avgFuncAtt(0, 0) - avgFuncDef(0, 0)
        else:
          arrayAvgs[i,j] = avgFuncAtt(T[i], k[j]) - avgFuncDef(T[i], k[j])
  arrayAvgs = transpose(arrayAvgs) # Transpose in order to plot T on x-axis and k on y-axis.

  # Create the heatmap
  norm = plt.Normalize(sclEndBtm, sclEndTop)

  g1 = sns.heatmap(arrayAvgs, cmap="YlOrRd",cbar=True, norm=norm)
  g1.set_ylabel('Production pay-off k')
  g1.set_xlabel('Production threshold T') 
  g1.invert_yaxis()

  g1.tick_params(axis='x', labelsize=9, rotation= 50)
  g1.tick_params(axis='y', labelsize=9, rotation= 0)

  plt.suptitle(title, verticalalignment='bottom')
  plt.tight_layout()



## FUNCTIONS FOR THE IMPLEMENTED VERSION OF THE PAD-C


### PAY-OFF FUNCTIONS

# Defender conflict pay-off function for the PAD-CImplemented.
def poDefI(a, d, eDef):  
    if a > d:
        return 0
    else:
        return (eDef + lDefI(d))

# Attacker conflict pay-off function for the PAD-CImplemented.
def poAttI(a, d, eAtt, eDef):  
    if a > d:
        return (eAtt + lAttI(a)) + (eDef  + lDefI(d))
    else:
        return (eAtt + lAttI(a))

# Defender production function PAD-CImplemented
def lDefI(d):
  if d == 0:
    return 33.484 - 20
  elif d == 1:
    return 32.484 - 20
  elif d == 2:
    return 31.484 - 20
  elif d == 3:
    return 30.484 - 20
  elif d == 4:
    return 29.484 - 20
  elif d == 5:    
    return 27.984 - 20
  elif d == 6: 
    return 24.486 - 20
  elif d == 7:    
    return 18.486 - 20
  elif d == 8:
    return 12 - 20
  elif d == 9:  
    return 11 - 20
  elif d == 10: 
    return 10 - 20
  elif d == 11: 
    return 9 - 20
  elif d == 12:
    return 8 - 20   
  elif d == 13: 
    return 7 - 20
  elif d == 14:     
    return 6 - 20
  elif d == 15: 
    return 5 - 20
  elif d == 16:
    return 4 - 20
  elif d == 17:   
    return 3 - 20
  elif d == 18:    
    return 2 - 20
  elif d == 19: 
    return 1 - 20
  elif d == 20:    
    return 0 - 20
  else:
    print("Out of range.")

# Defender production function PAD-CImplemented
# Return values based on Pay-off tables
# Subtract 20 because function returns max pay-off after investing 20 MU in optimal allocation choice of production and keep (with "multiplicator" x 1)
def lAttI(a):
  if a == 0:
    return 33.484 - 20
  elif a == 1:
    return 32.484 - 20
  elif a == 2:
    return 31.484 - 20
  elif a == 3:
    return 30.484 - 20
  elif a == 4:
    return 29.484 - 20
  elif a == 5:    
    return 27.984 - 20
  elif a == 6: 
    return 24.486 - 20
  elif a == 7:    
    return 18.486 - 20
  elif a == 8:
    return 12 - 20
  elif a == 9:  
    return 11 - 20 
  elif a == 10: 
    return 10 - 20 
  elif a == 11: 
    return 9 - 20 
  elif a == 12:
    return 8 - 20   
  elif a == 13: 
    return 7 - 20 
  elif a == 14:     
    return 6 - 20
  elif a == 15: 
    return 5 - 20
  elif a == 16:
    return 4 - 20
  elif a == 17:   
    return 3 - 20
  elif a == 18:    
    return 2 - 20
  elif a == 19: 
    return 1 - 20
  elif a == 20:    
    return 0 - 20
  else:
    print("Out of range.")


### OTHER FUNCTIONS for Implemented version

# Function to create a plot with with equilibrium predictions for the implemented version
def PADCEquilibPlotI():
  PODmatI = createDefPOMatPADCI()
  POAmatI = createAttPOMatPADCI()
  POADmatI = np.dstack((POAmatI,PODmatI))
  g_POADmatI = gt.NormalFormGame(POADmatI)
  PADCEquilibArr = gt.lemke_howson(g_POADmatI)
  # Create table with equilibrium strategies
  PADCEquilibTable = pd.DataFrame({'Strategy': list(range(0, eAtt+1)),
                               'Att Probability': PADCEquilibArr[0],
                               'Def Probability': PADCEquilibArr[1]})
  # Create plot with equilibrium strategies
  fig, (ax1,ax2) = plt.subplots(1,2, figsize=(12,5))  # 1 row, 2 columns
  PADCEquilibTable.plot.bar(x='Strategy', y=['Att Probability'], color=['Red'], ylim=(0,1), width=0.8, rot=50, ax=ax1)
  PADCEquilibTable.plot.bar(x='Strategy', y=['Def Probability'], color=['Blue'], ylim=(0,1), width=0.8, rot=50, ax=ax2)
  plt.suptitle('Equilibrium play for attacker and defender in the PAD-C', verticalalignment='bottom')
  ax2.set_yticklabels([],minor=False)
  plt.tight_layout()  # Optional ... often improves the lay

# Function to create Attacker Pay-off Matrix for implemented version
def createAttPOMatPADCI():
    eDef = 20
    eAtt =  20
    POAmat = np.zeros((eAtt+1, eDef+1),dtype=float)
    for i in range(eAtt+1):
        for j in range(eDef+1):
            POAmat[i,j]=poAttI(i, j, eAtt, eDef)
    return POAmat

# Function to create Defender Pay-off Matrix for implemented version
def createDefPOMatPADCI():
    eDef =  20
    PODmat = np.zeros((eDef+1,eAtt+1),dtype=float)
    for i in range(eDef+1):
        for j in range(eAtt+1):
            PODmat[i,j]=poDefI(i, j, eDef)
    return PODmat

# Function to create a table with equilibrium predictions for the implemented version
def PADCEquilibTableI():
  PODmatI = createDefPOMatPADCI()
  POAmatI = createAttPOMatPADCI()
  POADmatI = np.dstack((POAmatI,PODmatI))
  g_POADmatI = gt.NormalFormGame(POADmatI)
  PADCEquilibArr = gt.lemke_howson(g_POADmatI)
  # Create table with equilibrium strategies
  PADCEquilibTable = pd.DataFrame({'Strategy': list(range(0, eAtt+1)),
                               'Att Probability': PADCEquilibArr[0],
                               'Def Probability': PADCEquilibArr[1]})
  return PADCEquilibTable 


## Functions that compute additional variables based on a table *df* with columns for strategy *str* and probability *p_weight*

# A function to give a weighted average to get the average equilibrium investment
def wavg(df, str, p_weight):
    d = df[str]
    w = df[p_weight]
    return (d * w).sum()

# A function that returns the investment frequency.
def invFreq(df, str, p_weight):
    d = df.loc[df[str] == 0, p_weight]
    d = float(d)
    return 1 - d

# A function to give the conflict force under equilibrium. 
def invForce(df, str, p_weight):
    df = df.drop(0)
    strategies = df[str]
    pweights = df[p_weight]
    return (strategies * pweights).sum() / pweights.sum() 

# A function that returns the investment frequency for production.
def invFreqProd(df, str, p_weight):
    d = df.loc[df[str] == 0, p_weight]
    d = d.sum()
    return 1 - d
```
